```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap


# Set center coordinates for our target area (Lagos)
center_lat = 6.5244
center_lon = 3.3792


np.random.seed(42) 
num_points = 5000

lats = np.random.normal(loc=center_lat, scale=0.03, size=num_points)
lons = np.random.normal(loc=center_lon, scale=0.04, size=num_points)


df_routes = pd.DataFrame({'Latitude': lats, 'Longitude': lons})

city_map = folium.Map(location=[center_lat, center_lon], zoom_start=12, tiles='CartoDB dark_matter')

heat_data = [[row['Latitude'], row['Longitude']] for index, row in df_routes.iterrows()]

# HeatMap layer
HeatMap(
    heat_data,
    radius=25,       
    blur=15,         
    max_zoom=1,      
    gradient={0.4: 'blue', 0.65: 'lime', 1: 'red'}
).add_to(city_map)


output_file = "city_infrastructure_heatmap2.html"
city_map.save(output_file)
```
